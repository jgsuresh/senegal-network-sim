```python
# Mosquito bites person with 4 infections, each of which has its own probability of transmission
strain_ids = [1,2,3,4]
infectiousness = np.array([0.001,0.01,0.05,0.1])
# infectiousness = np.array([0.1,0.1,0.1,0.1])
```

```python
def fake_oocyst_formation(parent_strains, parent_strain_weights):
    # Choose 2 parent strains based on weights
    parent_strains = np.random.choice(parent_strains, 2, p=parent_strain_weights)
```

```python
diversity_counter = 0
for _ in range(100000):
    # Scenario 1: Each infection is picked up independently based on its own probability of transmission
    picked_up_by_mosquito = np.random.rand(4) < infectiousness
    # print(picked_up_by_mosquito)
    strains_in_mosquito = np.array(strain_ids)[picked_up_by_mosquito]
    # print(strains_in_mosquito)
    
    if np.sum(picked_up_by_mosquito) == 0:
        # print("No infections picked up by mosquito")
        pass
    else:
        strain_weights_in_mosquito = infectiousness[picked_up_by_mosquito]/np.sum(infectiousness[picked_up_by_mosquito])
        oocyst_parents = np.random.choice(strains_in_mosquito, 2, p=strain_weights_in_mosquito)
        
        if oocyst_parents[0] != oocyst_parents[1]:
            diversity_counter += 1
            print("Diversity!")
print(diversity_counter)
```

```python
diversity_counter = 0
for _ in range(100000):
    # Scenario 2: If any infection is picked up, all of them are picked up. 
    aggregate_infectiousness = 1-np.prod(1-infectiousness)
    picked_up_by_mosquito = np.random.rand() < aggregate_infectiousness
    
    if not picked_up_by_mosquito:
        # print("No infections picked up by mosquito")
        pass
    else:
        strains_in_mosquito = strain_ids
        strain_weights_in_mosquito = infectiousness/np.sum(infectiousness)
        oocyst_parents = np.random.choice(strains_in_mosquito, 2, p=strain_weights_in_mosquito)
        
        if oocyst_parents[0] != oocyst_parents[1]:
            diversity_counter += 1
            print("Diversity!")
print(diversity_counter)
```
