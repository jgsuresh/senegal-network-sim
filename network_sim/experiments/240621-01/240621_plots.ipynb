```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid")
sns.set_context("talk")
```

```python
df2 = pd.read_csv("../240620-02/sweep_data.csv")  # "Uniform biting, immunity ON"
df4 = pd.read_csv("../240620-04/sweep_data.csv") # "Heterogeneous (gamma=1) biting, immunity ON"
df6 = pd.read_csv("../240620-07/sweep_data.csv") # gamma=0.71 biting, immunity ON
df7 = pd.read_csv("../240620-08/sweep_data.csv") # gamma=2.4 biting, immunity ON
```

```python
df2.sort_values("prev", inplace=True)
df4.sort_values("prev", inplace=True)
df6.sort_values("prev", inplace=True)
df7.sort_values("prev", inplace=True)
```

```python
plt.figure(dpi=200)
plt.plot(df2["prev"], df2["mean_coi"], marker="o", label="Uniform biting", color="C1")
plt.plot(df4["prev"], df4["mean_coi"], marker="o", label="Jinja", color="C0")
plt.plot(df6["prev"], df6["mean_coi"], marker="o", label="Kanungu", color="C2")
plt.plot(df7["prev"], df7["mean_coi"], marker="o", label="Tororo", color="C3")
plt.xlabel("Prevalence")
plt.ylabel("Mean COI")
plt.legend(fontsize=10)
plt.tight_layout()
plt.ylim([0,10])
plt.savefig("coi_vs_prev.png")
```

```python
plt.figure(dpi=200)
plt.plot(df2["prev"], df2["polygenomic_frac"], marker="o", label="Uniform biting", color="C1")
plt.plot(df4["prev"], df4["polygenomic_frac"], marker="o", label="Jinja", color="C0")
plt.plot(df6["prev"], df6["polygenomic_frac"], marker="o", label="Kanungu", color="C2")
plt.plot(df7["prev"], df7["polygenomic_frac"], marker="o", label="Tororo", color="C3")
plt.xlabel("Prevalence")
plt.ylabel("Polygenomic Fraction")
plt.legend(fontsize=10)
plt.tight_layout()
# plt.ylim([0,10])
plt.savefig("poly_vs_prev.png")
```

```python
plt.figure(dpi=200)
plt.plot(df2["prev"], df2["mean_within_host_ibs"], marker="o", label="Uniform biting", color="C1")
plt.plot(df4["prev"], df4["mean_within_host_ibs"], marker="o", label="Jinja", color="C0")
plt.plot(df6["prev"], df6["mean_within_host_ibs"], marker="o", label="Kanungu", color="C2")
plt.plot(df7["prev"], df7["mean_within_host_ibs"], marker="o", label="Tororo", color="C3")
plt.xlabel("Prevalence")
plt.ylabel("Polygenomic Fraction")
plt.ylabel("Mean within-host IBS")
plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig("within_host_ibs_vs_prev.png")
```

```python
plt.figure(dpi=200)
plt.plot(df2["prev"], df2["proportion_unique"], marker="o", label="Uniform biting", color="C1")
plt.plot(df4["prev"], df4["proportion_unique"], marker="o", label="Jinja", color="C0")
plt.plot(df6["prev"], df6["proportion_unique"], marker="o", label="Kanungu", color="C2")
plt.plot(df7["prev"], df7["proportion_unique"], marker="o", label="Tororo", color="C3")
plt.xlabel("Prevalence")
plt.ylabel("Proportion unique (clonality)")
plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig("proportion_unique_vs_prev.png")
```
