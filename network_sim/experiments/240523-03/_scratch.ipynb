```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
```

```python
df = pd.read_csv("coi_prev.csv")
df2 = pd.read_csv("../240523-02/coi_prev.csv")
```

```python
# Sort dataframes by avg_prev
df = df.sort_values("avg_prev")
df2 = df2.sort_values("avg_prev")
```

```python
# Markers are circles connected by lines
plt.plot(df["avg_prev"], df["avg_COI"], label="Uniform", marker="o")
plt.plot(df2["avg_prev"], df2["avg_COI"], label="Heterogeneous (exponential)", marker="o")
plt.axhline(y=1, color='gray', linestyle='--')
plt.legend()
plt.xlabel("Prevalence")
plt.ylabel("Average COI")
```
