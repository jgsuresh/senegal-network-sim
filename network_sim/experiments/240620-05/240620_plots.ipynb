```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid")
sns.set_context("talk")
```

```python
df1 = pd.read_csv("../240620-01/sweep_data.csv")
df2 = pd.read_csv("../240620-02/sweep_data.csv")
df3 = pd.read_csv("../240620-03/sweep_data.csv")
df4 = pd.read_csv("../240620-04/sweep_data.csv")
```

```python
df1.columns
```

```python
df1.sort_values("prev", inplace=True)
df2.sort_values("prev", inplace=True)
df3.sort_values("prev", inplace=True)
df4.sort_values("prev", inplace=True)
```

```python
plt.figure(dpi=200)
plt.plot(df1["prev"], df1["mean_coi"], marker="o", label="Uniform biting, immunity OFF", color="C0", linestyle='dashed')
plt.plot(df2["prev"], df2["mean_coi"], marker="o", label="Uniform biting, immunity ON", color="C0")
plt.plot(df3["prev"], df3["mean_coi"], marker="o", label="Heterogeneous biting, immunity OFF", color="C1", linestyle='dashed')
plt.plot(df4["prev"], df4["mean_coi"], marker="o", label="Heterogeneous biting, immunity ON", color="C1")
plt.xlabel("Prevalence")
plt.ylabel("Mean COI")
plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig("coi_vs_prev.png")
```

```python
plt.figure(dpi=200)
plt.plot(df1["prev"], df1["polygenomic_frac"], marker="o", label="Uniform biting, immunity OFF", color="C0", linestyle='dashed')
plt.plot(df2["prev"], df2["polygenomic_frac"], marker="o", label="Uniform biting, immunity ON", color="C0")
plt.plot(df3["prev"], df3["polygenomic_frac"], marker="o", label="Heterogeneous biting, immunity OFF", color="C1", linestyle='dashed')
plt.plot(df4["prev"], df4["polygenomic_frac"], marker="o", label="Heterogeneous biting, immunity ON", color="C1")
plt.xlabel("Prevalence")
plt.ylabel("Polygenomic Fraction")
plt.ylim([0,1])
plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig("poly_vs_prev.png")
```

```python
plt.figure(dpi=200)
plt.plot(df1["prev"], df1["mean_within_host_ibs"], marker="o", label="Uniform biting, immunity OFF", color="C0", linestyle='dashed')
plt.plot(df2["prev"], df2["mean_within_host_ibs"], marker="o", label="Uniform biting, immunity ON", color="C0")
plt.plot(df3["prev"], df3["mean_within_host_ibs"], marker="o", label="Heterogeneous biting, immunity OFF", color="C1", linestyle='dashed')
plt.plot(df4["prev"], df4["mean_within_host_ibs"], marker="o", label="Heterogeneous biting, immunity ON", color="C1")
plt.xlabel("Prevalence")
plt.ylabel("Mean within-host IBS")
plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig("within_host_ibs_vs_prev.png")
```

```python
plt.figure(dpi=200)
plt.plot(df1["prev"], df1["proportion_unique"], marker="o", label="Uniform biting, immunity OFF", color="C0", linestyle='dashed')
plt.plot(df2["prev"], df2["proportion_unique"], marker="o", label="Uniform biting, immunity ON", color="C0")
plt.plot(df3["prev"], df3["proportion_unique"], marker="o", label="Heterogeneous biting, immunity OFF", color="C1", linestyle='dashed')
plt.plot(df4["prev"], df4["proportion_unique"], marker="o", label="Heterogeneous biting, immunity ON", color="C1")
plt.xlabel("Prevalence")
plt.ylabel("Proportion unique (clonality)")
plt.legend(fontsize=10)
plt.tight_layout()
plt.savefig("proportion_unique_vs_prev.png")
```
