```python
# Draw biting risk from an exponential distribution
import numpy as np
import matplotlib.pyplot as plt

# Draw 1000 samples from an exponential distribution with a mean of 1
samples = np.random.exponential(1, 10000)
```

```python
plt.hist(samples, bins=100)
```

```python
# Calculate the standard deviation of the samples
np.std(samples)
```

```python
# Calculate the integral of the distribution for different quartiles of the distribution
```

```python
# Calculate the integral of the distribution for the first quartile
np.mean(samples[samples < np.percentile(samples, 25)])
```

```python
# Calculate the integral of the distribution for the second quartile
np.mean(samples[(samples >= np.percentile(samples, 25)) & (samples < np.percentile(samples, 50))])
```

```python
# Calculate the integral of the distribution for the third quartile
np.mean(samples[(samples >= np.percentile(samples, 50)) & (samples < np.percentile(samples, 75))])
```

```python
# Calculate the integral of the distribution for the fourth quartile
np.mean(samples[samples >= np.percentile(samples, 75)])
```

```python
# Sort samples and calculate the cumulative sum
sorted_samples = np.sort(samples)
cumulative_sum = np.cumsum(sorted_samples)

# Get rank of each sample
rank = np.arange(1, len(samples) + 1)
normalized_rank = rank / len(samples)
```

```python
plt.plot(normalized_rank, cumulative_sum/np.max(cumulative_sum))
plt.xlabel('Normalized rank')
plt.ylabel('Cumulative sum')
```

```python
# Find point where the cumulative sum reaches 50% of the total sum
threshold = 0.5
index = np.argmax(cumulative_sum / np.max(cumulative_sum) > threshold)
normalized_rank[index]
```

```python
# Find point where the cumulative sum reaches 80% of the total sum
threshold = 0.8
index = np.argmax(cumulative_sum / np.max(cumulative_sum) > threshold)
normalized_rank[index]
```

```python
# Find point where the cumulative sum reaches 50% of the total sum
for threshold in [0.25, 0.5, 0.75]:
    index = np.argmax(cumulative_sum / np.max(cumulative_sum) > threshold)
    print(normalized_rank[index])
```
