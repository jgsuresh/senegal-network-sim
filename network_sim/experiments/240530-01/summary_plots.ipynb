```python
# Load data
df1 = pd.read_csv("coi_prev.csv")
df2 = pd.read_csv("pop_ibd.csv")
df3 = pd.read_csv("pop_ibs.csv")
df4 = pd.read_csv("polygenomic_fraction.csv")
df5 = pd.read_csv("barcode_frequencies.csv")
```

```python
df = (df1.merge(df2.rename(columns={"ibs": "ibd"}), on="sim_id")
      .merge(df3, on="sim_id")
      .merge(df4, on="sim_id")
      .merge(df5, on="sim_id")
      .sort_values("avg_prev")
      .reset_index(drop=True))
```

```python
plt.scatter(df["avg_prev"], df["avg_COI"])
plt.xlim([0,1])
plt.xlabel("Prevalence")
plt.ylabel("Average COI")
```

```python
plt.scatter(df["avg_prev"], df["ibd"])
plt.ylim([0,0.1])
plt.xlabel("Prevalence")
plt.ylabel("Population IBD after 3 years")
plt.title("Level 0 complexity (uniform biting)")
plt.ylim([0,0.04])
```

```python
plt.scatter(df["avg_prev"], df["ibs"])
plt.ylim([0.5,0.53])
plt.xlabel("Prevalence")
plt.ylabel("Population IBS after 3 years")
```

```python
plt.scatter(df["avg_prev"], df["polygenic_fraction"])
plt.ylim([0,1])
plt.xlim([0,1])
plt.xlabel("Prevalence")
plt.ylabel("Polygenomic fraction at the end of 3 years")
```
